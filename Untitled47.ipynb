```python

l = ['s1','s2','s3','m1','m2','m3','l1','l2','l3']
d = dict.fromkeys(l," ")

    
def check():
    # horizontal check
    if((d['s1'] == 'X' and d['s2']=='X' and d['s3'] == 'X') or (d['m1'] == 'X' and d['m2']=='X' and d['m3'] == 'X') or
       (d['l1'] == 'X' and d['l2']=='X' and d['l3'] == 'X') or (d['s1'] == 'X' and d['m2']=='X' and d['l3'] == 'X') or 
       (d['s3'] == 'X' and d['m2']=='X' and d['l1'] == 'X') or (d['s1'] == 'X' and d['m1']=='X' and d['l1'] == 'X') or 
       (d['s2'] == 'X' and d['m2']=='X' and d['l2'] == 'X') or (d['s3'] == 'X' and d['m3']=='X' and d['l3'] == 'X')):
        print('Player_1 won !')
        return 1
    elif((d['s1'] == 'O' and d['s2']=='O' and d['s3'] == 'O') or (d['m1'] == 'O' and d['m2']=='O' and d['m3'] == 'O') or
       (d['l1'] == 'O' and d['l2']=='O' and d['l3'] == 'O') or (d['s1'] == 'O' and d['m2']=='O' and d['l3'] == 'O') or 
       (d['s3'] == 'O' and d['m2']=='O' and d['l1'] == 'O') or (d['s1'] == 'O' and d['m1']=='O' and d['l1'] == 'O') or 
       (d['s2'] == 'O' and d['m2']=='O' and d['l2'] == 'O') or (d['s3'] == 'O' and d['m3']=='O' and d['l3'] == 'O')):
        print('Player_2 won !')
        return 1
    else:
        return 0


        
```

```python
player = 1
total_moves = 0
end_check = 0
while True:
    print(d['s1']+'|'+d['s2']+'|'+d['s3'])
    print('-+-+-')
    print(d['m1']+'|'+d['m2']+'|'+d['m3'])
    print('-+-+-')
    print(d['l1']+'|'+d['l2']+'|'+d['l3'])
    
    end_check = check()
    if total_moves == 9 or end_check ==1:
        break
    while True:
        if player ==1:
            p1_input = input('player 1 enter your position for X')
            if p1_input in d and d[p1_input] == " ":
                d[p1_input] = 'X'
                player = 2
                break
            else:
                print('Invalid input plz try once again')
                continue
        else:
            p2_input = input('player 2 enter position for O')
            if p2_input in d and d[p2_input] == " ":
                d[p2_input] = 'O'
                player = 1
                break
            else:
                print('Invalid input plz try once again')
                continue
        total_moves += 1
        print("+++++++++++++++++++++++++++++")
        print()
```
